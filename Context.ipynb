```python
from reactpy import Reactive
import numpy as np
import pandas as pd
```

```python
r = Reactive()
r.a = 2
r.b = r(lambda a: a*a)
before = r.b
with r.context(a=5):
    during = r.b
after = r.b
before, during, after
```

```python
r = Reactive()
r.c = r(lambda a,b: abs(a-b))
r.result = r(lambda a,b,c: a*b + c)

data = {}
for a, b in ((a, b) for a in range(10) 
                    for b in range(10)):
    with r.context(a=a, b=b):
        data[(r.a, r.b)] = r.result 
```

```python
pd.Series(data).unstack()
```
